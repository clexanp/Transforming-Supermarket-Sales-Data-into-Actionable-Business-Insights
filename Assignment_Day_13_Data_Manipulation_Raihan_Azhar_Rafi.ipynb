```python
data_penjualan = pd.read_csv('/content/drive/MyDrive/Assignment Day 13 Data Manipulation_Raihan Azhar Rafi/supermarket_sales.csv')
data_penjualan.head(10)
```

```python
# Langkah pertama kita mau melakukan proses filtering dengan cara mengambil setiap transaksi dari customer yang hanya memakai
# metode pembayaran dengan E-Wallet.
transaksi_ewallet = data_penjualan[data_penjualan['Payment'] == 'Ewallet']

# Langkah kedua kita mau melakukan proses filtering terhadap setiap transaksi dari customer yang tidak memakai metode pembayaran
# dengan E-Wallet.
transaksi_bukan_ewallet = data_penjualan[data_penjualan['Payment'] != 'Ewallet']

# Langkah ketiga kita melakukan pengambilan jumlah data transaksi yang dilakukan oleh customer menggunakan metode E-Wallet.
jumlah_transaksi_ewallet = transaksi_ewallet.shape[0]

# Langkah keempat kita melakukan perhitungan rata-rata terhadap metode pembayaran E-Wallet yang dipakai customer untuk bertransaksi.
mean_rating_ewallet = transaksi_ewallet['Rating'].mean()

# Langkah kelima kita melakukan perhitungan rata-rata terhadap metode pembayaran yang tidak menggunakan E-Wallet yang dipakai customer
# untuk bertransaksi.
mean_rating_bukan_ewallet = transaksi_bukan_ewallet['Rating'].mean()

# Langkah keenam kita melakukan perhitungan rata-rata rating keseluruhan dari dataset.
mean_rating_semua = data_penjualan['Rating'].mean()

# Langkah ketujuh kita mau menampilkan hasil analisisnya.
print("Proses Analisis:")
print(f"Jumlah Transaksi Yang Memakai E-Wallet: {jumlah_transaksi_ewallet:,} Transaksi")
print(f"Rata-Rata Rating Yang Memakai E-Wallet: {mean_rating_ewallet:.2f}")
print(f"Rata-Rata Rating Yang Tidak Memakai E-Wallet: {mean_rating_bukan_ewallet:.2f}")
print(f"Rata-Rata Rating Keseluruhan: {mean_rating_semua:.2f}")

print("\nHasil Analisis:")
if mean_rating_ewallet > mean_rating_bukan_ewallet:
  print(f"Customer E-Wallet Memiliki Rating Yang Lebih Tinggi ({mean_rating_ewallet - mean_rating_bukan_ewallet:.2f} poin lebih tinggi)")
elif mean_rating_ewallet < mean_rating_bukan_ewallet:
  print(f"Customer E-Wallet Memiliki Rating Yang Lebih Rendah ({mean_rating_bukan_ewallet - mean_rating_ewallet:.2f} poin lebih rendah)")
else:
  print("Tidak Perbedaan Signifikan Pada Rating Antara Customer E-Wallet dengan Customer Non E-Wallet")
```

# Jawaban Soal 2: Coding Untuk Metode Sorting

```python
data_penjualan = pd.read_csv('/content/drive/MyDrive/Assignment Day 13 Data Manipulation_Raihan Azhar Rafi/supermarket_sales.csv')
data_penjualan.head(10)
```

```python
# Langkah pertama kita mau mengkonversi kolom 'Date' ke dalam format DateTime agar bisa di ekstrak datanya berdasarkan bulan dan tahun.
data_penjualan['Date'] = pd.to_datetime(data_penjualan['Date'])

# Langkah kedua kita mau ekstrak ke dalam periode bulan dan tahun dalam format YYYY-MM supaya nantinya bisa di grouping per bulan datanya.
data_penjualan['Bulan'] = data_penjualan['Date'].dt.to_period('M')

# Langkah ketiga kita mau filter data customer hanya pria dan dua jenis produk yang sedang dianalisis (Health and Beauty) dan (Sports and Travel).
produk_pilihan_pria = ['Health and beauty', 'Sports and travel']
data_filter_pria = data_penjualan[(data_penjualan['Gender'] == 'Male') &
                             (data_penjualan['Product line'].isin(produk_pilihan_pria))]

# Langkah keempat kita mau melakukan grouping untuk menghitung total penjualan per bulan dan juga untuk total penjualan per lini produk.
penjualan_per_bulan = data_filter_pria.groupby(['Bulan', 'Product line'])['Total'].sum().reset_index()

# Langkah kelima kita mau melakukan sorting untuk mengurutkan data berdasarkan bulan secara ascending (dari yang terkecil ke terbesar)
# dan untuk total penjualan secara descending (dari yang terbesar ke terkecil).
sorting_penjualan = penjualan_per_bulan.sort_values(by=['Bulan', 'Total'], ascending=[True, False])

# Langkah keenam kita mau menampilkan hasil dari proses sorting datanya.
print("Tabel Analisis : \n")
print(sorting_penjualan.to_string(index=False))

# Langkah ketujuh kita mau menghitung total keseluruhan untuk referensi perbandingan antara produk (Health and Beauty) dan (Sports and Travel).
total_health_beauty = penjualan_per_bulan[penjualan_per_bulan['Product line'] == 'Health and beauty']['Total'].sum()
total_sports_travel = penjualan_per_bulan[penjualan_per_bulan['Product line'] == 'Sports and travel']['Total'].sum()

# Langkah kedelapan kita mau menampilkan hasil analisisnya.
print("\nHasil Analisis Minat Customer Pria Terhadap Produk (Health and Beauty) VS (Sports and Travel) Dalam Menentukan Kegiatan Self Reward: ")
print(f"Total penjualan untuk produk (Health and Beauty): {total_health_beauty:,.0f}")
print(f"Total penjualan untuk produk (Sports and Travel): {total_sports_travel:,.0f}")

if total_sports_travel > total_health_beauty:
    print("Customer Pria Lebih Banyak Membeli Produk (Sports and Travel")
elif total_health_beauty > total_sports_travel:
    print("Customer Pria Lebih Banyak Membeli Produk (Health and Beauty)")
else:
    print("Pemilihan Produk (Health and Beauty) Dan Produk (Sports and Travel) Seimbang")
```

# Jawaban Soal 4: Coding Untuk Metode Join / Merging

```python
data_penjualan = pd.read_csv('/content/drive/MyDrive/Assignment Day 13 Data Manipulation_Raihan Azhar Rafi/supermarket_sales.csv')
data_penjualan.head(10)
```

```python
data_cabang = pd.read_csv('/content/drive/MyDrive/Assignment Day 13 Data Manipulation_Raihan Azhar Rafi/branch_info.csv')
data_cabang.head()
```

```python
data_produk = pd.read_csv('/content/drive/MyDrive/Assignment Day 13 Data Manipulation_Raihan Azhar Rafi/product_info.csv')
data_produk.head()
```

```python
# Langkah pertama kita mau melakukan merging data dengan menambahkan informasi dari kolom region pada DataFrame branch_info.
data_region = pd.merge(data_penjualan, data_cabang, on='Branch', how='left')

# Langkah kedua kita mau melanjutkan merging data sebelumnya dengan menambahkan informasi dari kolom Product line pada DataFrame product_info.
data_komplet = pd.merge(data_region, data_produk, on='Product line', how='left')

# Langkah ketiga kita mau melakukan grouping data berdasarkan Customer type, Branch, Region, dan Product line.
grouping_kolom = data_komplet.groupby(['Customer type', 'Branch', 'Region', 'Product line']).agg({
    'Total'      : 'sum',  # Total pengeluaran transaksi customer
    'Rating'     : 'mean', # Rata-rata rating customer
    'Invoice ID' : 'count' # Jumlah transaksi customer
}).reset_index()

# Langkah keempat kita mau meratakan nama kolom dari dataset yang sudah di agregasi ke dalam penamaan bahasa indonesia supaya lebih mudah dibacanya.
grouping_kolom.columns = ['Tipe_Customer', 'Cabang', 'Region', 'Product_Line',
                             'Total_Pengeluaran', 'Rata_Rata_Rating', 'Jumlah_Transaksi']

# Langkah kelima kita mau melakukan sorting untuk mengurutkan kolom Region secara ascending dan untuk kolom Total_Pengeluaran secara descending.
sorting_kolom = grouping_kolom.sort_values(by=['Region', 'Total_Pengeluaran'], ascending=[True, False])

# Langkah keenam kita mau menampilkan isi DataFrame dari sorting_kolom tanpa nomor index (index=False) dan menjadikan outputnya seperti tabel.
print("Tabel Analisis:\n")
print(sorting_kolom.to_string(index=False))

# Langkah ketujuh kita mau menampilkan hasil analisisnya.
print("\nHasil Analisis:")
data_member = sorting_kolom[sorting_kolom['Tipe_Customer'] == 'Member']
data_normal = sorting_kolom[sorting_kolom['Tipe_Customer'] == 'Normal']

if not data_member.empty and not data_normal.empty:
    if data_member['Total_Pengeluaran'].mean() > data_normal['Total_Pengeluaran'].mean():
        print("Customer Member secara keseluruhan memiliki total pengeluaran lebih tinggi dibandingkan Customer Normal di berbagai region/cabang.")
    else:
        print("Customer Normal secara keseluruhan memiliki total pengeluaran lebih tinggi dibandingkan Customer Member di berbagai region/cabang.")
    if data_member['Rata_Rata_Rating'].mean() > data_normal['Rata_Rata_Rating'].mean():
        print("Customer Member memiliki rata-rata rating lebih tinggi dibandingkan dengan Customer Normal di berbagai region/cabang.")
    else:
        print("Customer Normal memiliki rata-rata rating lebih tinggi dibandingkan dengan Customer Member di berbagai region/cabang.")
```

Aku mengerjakannya sesuai ketentuan konteks soal dengan memfilter data hanya untuk 1 bulan terakhir dan mengekstrak data bulan dari kolom Date, selanjutnya membuat pivot table dengan Product line sebagai baris, Branch sebagai kolom, dan Quantity sebagai nilai fungsi agregasi sum.

# Jawaban Soal 5: Coding Untuk Metode Pivot

```python
data_penjualan = pd.read_csv('/content/drive/MyDrive/Assignment Day 13 Data Manipulation_Raihan Azhar Rafi/supermarket_sales.csv')
data_penjualan.head(10)
```

```python
# Langkah pertama kita mau melakukan konversi terhadap kolom Date ke dalam format datetime untuk filter dan analisis waktu.
data_penjualan['Date'] = pd.to_datetime(data_penjualan['Date'])

# Langkah kedua kita mau melakukan pencarian tanggal maksimum untuk menentukan data dari 1 bulan terakhir.
tanggal_maksimum = data_penjualan['Date'].max()
tanggal_minimum_bulan_terakhir = tanggal_maksimum - pd.Timedelta(days=30)

# Langkah ketiga kita mau melakukan filtering data hanya untuk 1 bulan terakhir berdasarkan tanggal.
data_bulan_terakhir = data_penjualan[data_penjualan['Date'] >= tanggal_minimum_bulan_terakhir]

# Langkah keempat kita mau membuat tabel pivot untuk Sum Quantity (total kuantitas) dengan Product line sebagai baris dan Branch sebagai kolom.
tabel_pivot = pd.pivot_table(
    data_bulan_terakhir,
    values='Quantity',       # Kolom numerikal (Quantity untuk hitung total terjual)
    index='Product line',    # Kolom kategorikal (Product line sebagai baris)
    columns='Branch',        # Kolom kategorikal (Branch sebagai kolom)
    aggfunc='sum'            # Fungsi agregasi (sum untuk total quantity)
)

# Langkah kelima kita mau menampilkan hasil pivot tabelnya.
print("Pivot Table: Total Quantity per Product Line yang Terjual di Cabang Tertentu Selama (1 Bulan Terakhir)\n")
print(tabel_pivot)

# Langkah keenam kita mau menampilkan hasil analisisnya dengan mencari Product line paling laris di cabang tertentu.
print("\nHasil Analisis:")
for cabang in tabel_pivot.columns:
    produk_terlaris = tabel_pivot[cabang].idxmax()
    kuantitas_maksimum = tabel_pivot[cabang].max()
    print(f"Cabang {cabang}: Product line yang paling banyak terjual adalah '{produk_terlaris}' dengan total {kuantitas_maksimum} unit.")
```

```python
# Langkah pertama kita mau membuat tabel crosstab untuk mengukur frekuensi antara metode pembayaran (Payment) dengan cabang (Branch).
tabel_crosstab = pd.crosstab(
    index=data_penjualan['Payment'], # Parameter disamping ini menentukan bahwa metode pembayaran (Credit Card, E-Wallet, dan Cash) menjadi baris pada tabel.
    columns=data_penjualan['Branch'], # Parameter disamping ini menentukan bahwa cabang supermarket (A, B, C) menjadi kolom pada tabel.
    margins=True, # Parameter di samping ini memerintahkan Pandas menambahkan baris dan kolom di bagian bawah pada sisi kiri dan kanan tabel.
    margins_name='Total' # Parameter di samping ini untuk memberikan nama 'Total' pada baris dan kolom yang sudah ditambahkan sebelumnya.
)

# Langkah kedua kita mau menampilkan hasil tabel crosstab yang sebelumnya sudah dibuat.
print("Table Crosstab: Frekuensi Penggunaan Metode Pembayaran per Cabang\n")
print(tabel_crosstab)

# Langkah ketiga kita mau menghitung persentase dari metode pembayaran (Credit card) pada tiap cabang dengan cara mengambil baris 'Credit card' saja tanpa total.
proporsi_credit_card = tabel_crosstab.loc['Credit card'] / tabel_crosstab.loc['Total'] * 100

# Langkah keempat kita mau menghitung persentase dari metode pembayaran (E-Wallet).
proporsi_ewallet = tabel_crosstab.loc['Ewallet'] / tabel_crosstab.loc['Total'] * 100

# Langkah kelima kita mau menampilkan persentase dari (Credit card) dan (E-Wallet) pada tiap cabang.
print("\nTable Persentase Perbandingan (Credit Card) VS (E-Wallet):")
print("Persentase penggunaan Credit Card di setiap cabang:")
print(proporsi_credit_card.round(2).astype(str) + '%')

print("\nPersentase penggunaan E-Wallet di setiap cabang:")
print(proporsi_ewallet.round(2).astype(str) + '%')

# Langkah keenam kita mau menampilkan hasil analisis berdasarkan frekuensi penggunaan metode (Credit card).
cabang_tertinggi_credit = proporsi_credit_card.idxmax()
persentase_tertinggi_credit = proporsi_credit_card.max()

print(f"\nTable Hasil Analisis:")
print(f"Cabang dengan penggunaan Credit card paling tinggi adalah cabang {cabang_tertinggi_credit} dengan persentase sebesar {persentase_tertinggi_credit:.2f}%.")
```

# Jawaban Soal 8: Coding Untuk Metode Mengeksplor terkait dengan operasi pada Datetime (week, day, dan sebagainya)

```python
data_penjualan = pd.read_csv('/content/drive/MyDrive/Assignment Day 13 Data Manipulation_Raihan Azhar Rafi/supermarket_sales.csv')
data_penjualan.head(10)
```

```python
# Langkah pertama kita mau mengkonversi kolom Time ke dalam datetime.
data_penjualan['Time'] = pd.to_datetime(data_penjualan['Time'], format='%H:%M')

# Langkah kedua kita mau mengekstrak kolom Time menjadi jam harian.
data_penjualan['Jam'] = data_penjualan['Time'].dt.hour

# Langkah ketiga kita mau mengklompokkan menjadi periode waktu (Pagi, Siang, Sore).
def kelompok_periode(jam):
    if 9 <= jam < 13:
        return 'Pagi (09:00 - 12:59)'
    elif 13 <= jam < 17:
        return 'Siang (13:00 - 16:59)'
    elif 17 <= jam < 21:
        return 'Sore (17:00 - 20:59)'
    else:
        return 'Malam (21:00 - 08:59)'

data_penjualan['Periode_Waktu'] = data_penjualan['Jam'].apply(kelompok_periode)

# Langkah keempat kita mau menghitung total penjualan berdasarkan periode waktu harian dan Product line.
total_periode_produk = data_penjualan.groupby(['Periode_Waktu', 'Product line'])['Total'].sum().reset_index()

# Langkah kelima kita mau menghitung total penjualan per periode waktu harian untuk menemukan waktu tersibuk.
total_periode_tersibuk = total_periode_produk.groupby('Periode_Waktu')['Total'].sum().reset_index()
total_periode_tersibuk = total_periode_tersibuk.sort_values('Total', ascending=False)

# Langkah keenam kita mau mengurutkan periode waktu harian untuk nantinya ditampilkan datanya.
urutan_periode = ['Pagi (09:00 - 12:59)', 'Siang (13:00 - 16:59)',
                  'Sore (17:00 - 20:59)', 'Malam (21:00 - 08:59)']
total_periode_produk['Periode_Waktu'] = pd.Categorical(total_periode_produk['Periode_Waktu'],
                                                 categories=urutan_periode, ordered=True)
analisis_sorted = total_periode_produk.sort_values(['Periode_Waktu', 'Total'], ascending=[True, False])

# Langkah ketujuh kita mau menampilkan detail tabel terkait penjualan berdasarkan periode waktu harian dan total penjualan beserta list per Product line.
print("Tabel Analisis Penjualan per Periode Waktu dan Lini Produk\n")
print(analisis_sorted.to_string(index=False))

# Langkah kedelapan kita mau menampilkan hasi analisisnya dengan cara mencari Product line paling laris di periode waktu tersibuk.
print("\nHasil Analisis:")
if not total_periode_tersibuk.empty:
    periode_sibuk = total_periode_tersibuk.iloc[0]['Periode_Waktu']
    total_penjualan_sibuk = total_periode_tersibuk.iloc[0]['Total']

    sorting_periode_sibuk = analisis_sorted[analisis_sorted['Periode_Waktu'] == periode_sibuk]
    produk_dominan = sorting_periode_sibuk.iloc[0]['Product line']
    total_produk_laris = sorting_periode_sibuk.iloc[0]['Total']

    print(f"Periode waktu tersibuk adalah '{periode_sibuk}' dengan total penjualan: {total_penjualan_sibuk:.2f} unit.")
    print(f"Product line paling laris pada periode waktu tersebut adalah '{produk_dominan}' dengan total penjualan: {total_produk_laris:.2f} unit.")
```

```python
data_penjualan = pd.read_csv('/content/drive/MyDrive/Assignment Day 13 Data Manipulation_Raihan Azhar Rafi/supermarket_sales.csv')
data_penjualan.head(10)
```

```python
from sklearn.preprocessing import StandardScaler

# Langkah pertama kita mau melakukan pemilihan untuk 4 kolom numerik yang bisa untuk di scaling dan juga sesuai dengan konteks soal.
kolom_numerik = ['Total', 'cogs', 'gross income', 'Rating']

# Langkah kedua kita mau melakukan pembuatan DataFrame baru yang hanya diisi oleh 4 kolom numerik yang sudah dijelaskan sebelumnya.
data_numerik = data_penjualan[kolom_numerik].copy()

# Langkah ketiga kita mau melakukan inisialisasi objek StandardScaler dari scikit-learn.
scaler = StandardScaler()

# Langkah keempat kita mau menerapkan scaling pada 4 kolom numerik.
data_scaled = scaler.fit_transform(data_numerik)

# Langkah kelima kita mau mengubah hasil scaling menjadi DataFrame supaya mudah saat dibaca.
data_scaled_df = pd.DataFrame(data_scaled, columns=kolom_numerik)

# Langkah keenam kita mau menampilkan hasil tabel analisis untuk melihat perubahan data setelah proses scaling.
print("Tabel Analisis Hasil Scaling Dengan Standard Scaler (10 Baris Pertama):")
print(data_scaled_df.head(10).round(4))

# Langkah ketujuh kita mau menampilkan hasil analisis untuk deskriptif statistik setelah dilakukan proses scaling.
print("\nStatistik Deskriptif Setelah Scaling:")
print(data_scaled_df.describe().round(4))
```
